```python
from googletrans import Translator
```

```python
# 翻訳元ファイルと保存先ファイル
input_file = 'mc_dev_data.txt'      # 翻訳したいファイル
output_file = 'mc_dev_data_jp.txt'  # 翻訳結果を保存するファイル
```

```python
# 翻訳オブジェクトの作成
translator = Translator()
```

```python
# ファイルの読み込みと翻訳
with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
    for line in infile:
        line = line.strip()  # 行の前後の空白を削除
        if not line:         # 空行はスキップ
            continue
        
        # ラベルと本文を分割
        parts = line.split(' ', 1)  # 最初のスペースで分割
        if len(parts) < 2:
            continue  # ラベルや本文が欠けている場合はスキップ
        
        label, text = parts[0], parts[1]
        
        # 翻訳実行（英語 -> 日本語）
        translated = translator.translate(text, src='en', dest='ja')
        
        # 翻訳結果を保存（ラベルを元に戻す）
        outfile.write(f"{label} {translated.text}\n")

        # 進捗を表示
        print(f"Label: {label}")
        print(f"Original: {text}")
        print(f"Translated: {translated.text}")
        print("-" * 30)

print("翻訳が完了しました。結果は 'datasets/mc_dev_data_jp.txt' に保存されました。")
```
