```python
#googletransをインポート
from googletrans import Translator

# 翻訳するファイルと翻訳したファイルのPATH
input_file = 'mc_train_data.txt'      # 翻訳したいファイル
output_file = 'mc_train_data_jp.txt'  # 翻訳したファイル

#オブジェクトの指定
translator = Translator()

#infileで読み込み、outfileで書き込む
with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', encoding='utf-8') as outfile:
    for line in infile:
        #空白の削除
        line = line.strip()
        if not line:
            continue
        
        #空白で分割
        parts = line.split(' ', 1) 
        if len(parts) < 2:
            continue
        
        #分割したものをラベルとテキストに代入
        label, text = parts[0], parts[1]
        
        #テキストを日本語に翻訳
        translated = translator.translate(text, src='en', dest='ja')
        
        #ラベルと翻訳した文を書き込む
        outfile.write(f"{label} {translated.text}\n")

print("翻訳が完了しました。結果は 'datasets/mc_train_data_jp.txt' に保存されました。")
```
