```python
import spacy
from spacy import displacy

nlp = spacy.load("ja_ginza")
doc = nlp("彼は傘を持って出かけました。")
displacy.render(doc, style='dep')
```

```python
def excerpt(doc):
    elements = []
    for token in doc:
        if token.dep_ == "nsubj":
            #主語の名詞部分(私など)
            phrase = token.text
            #caseもくっつける(はなど)
            for child in token.children:
                if child.dep_ == "case":
                    phrase += child.text
            elements.append((phrase, "主語"))
            
        elif token.dep_ == "obj":
            phrase = token.text
            for child in token.children:
                if child.dep_ == "case":
                    phrase += child.text
            elements.append((phrase, "目的語"))
        elif token.dep_ == "ROOT":
            phrase = token.text
            #caseもくっつける(はなど)
            for child in token.children:
                if child.dep_ == "aux":
                    phrase += child.text
            elements.append((phrase, "述語"))
        else:
            pass  # 何もしない
    return elements
```

```python
doc = nlp("私はギリシャのピザを食べたいです。")
result = excerpt(doc)
print(result)
```

```python
new_sent = ''.join(word for word, role in result)
print(new_sent)
```
